## 2. Pie Chart ##

Alasan menggunakan Pie Chart karena yang pertama adalah data yang sedikit lalu pie chart juga dengan mudah menunjukkan berapa persen peluang dari data tersebut

setelah itu menurut data peluang pengambilan permen dalam kantong secara acak, Kopiko mendapatkan 15% peluang 

```python
import matplotlib.pyplot as plt

nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
Jumlah_permen = [52, 39, 78, 13, 78]
warna = ('#A0522D','#FF1493','#FFFACD','#7CFC00','#FF8C00')
highlight = (0,0.1,0,0,0)

plt.title("Pelunag Pengambilan Permen Dalam Kantong")
plt.pie(Jumlah_permen, labels= nama_permen, autopct = '%1.0f%%', colors= warna, explode = highlight, shadow = True)
plt.show
```

## 3. Bar Chart ##

alasan memilih barchart karena dalam bar chart kita dapat mengurutkan data dari yang terkecil ke terbesar begitu juga dengan sebaliknya maka dalam kasus ini akan terlihat manakah makanan penutup yang tidak populer  dan yang populer

makanan penutup yang tidak populer adalah 
- Puding Vanila(6)
- Pastel(5)
- Kue Wajik(2)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datapenjualan_makananpenutup = [
    'Donat','Pastel','Kue Coklat',
    'Ice Cream','Puding Vanila','Brownies',
    'Puding Strawberry','Puding Coklat','Ice Cream Nutela',
    'Kue Coklat-Keju','Kue Wajik','Kue Sus','Mochi']

Jumlah_makananpenutup = [14,5,12,19,6,8,12,9,10,17,2,9,13]

df = pd.DataFrame({'Makanan':datapenjualan_makananpenutup, 'Penjualan':Jumlah_makananpenutup})
df.sort_values(by='Penjualan', inplace= True)

koor_x = np.arange(len(df))
warna = ['#0000FF' for _ in range(len(df))]
warna [-11] = '#FF0000'
warna [-12] = '#FF0000'
warna [-13] = '#FF0000'

# plt.figure(figsize=(20,10))

plt.barh(koor_x, df['Penjualan'], tick_label=df['Makanan'],color= warna)
plt.xlabel('Jumlah Penjualan')
plt.title("Visualisasi Popularitas Relatif Makanan Penutup")
plt.show()
```
